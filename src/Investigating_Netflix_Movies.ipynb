```python
# Libary cell
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read file csv as DataFrame
# Đọc file csv thông qua pandas chuyển về kiểu DataFrame
file_path = r"D:\Repo-train\Jnotebook\Neflix_Project_DataCamp\datasets\netflix_data.csv"
netflix_df = pd.read_csv(file_path)
# Sample the first 10 rows to evaluate the data format
# Lấy 10 dòng bất kỳ nhìn tổng quan về dữ liệu
random_sample = netflix_df.sample(n=10)
print(random_sample)
# Ensure the relevant columns are available
# Kiểm tra các cột giá trị
print("\nDanh sách các cột trong DataFrame:")
print(netflix_df.columns)
```

```python
# Filter movies from the 1990s
# Lọc ra các phim trong thập niên 90s
netflix90s_df = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]
```

```python
# Type of duration
# Kiểm tra loại dữ liệu của thời lượng phim
print(netflix90s_df['duration'].dtype)
```

```python
# Convert duration to numeric (if it's not already) and handle missing values
# Chuyển đổi dữ liệu về một chuẩn nhất quán ( có thể kiểu hiện tại có những sai sót) và kiểm tra những hàng còn thiếu thành NaN (not a number) để quá trình tính toán không sai soát như các khác trịn N/N hay unknown
netflix90s_df.loc[:, 'duration'] = pd.to_numeric(netflix90s_df['duration'], errors='coerce')
durations = netflix90s_df['duration'].dropna().to_numpy()

# Find the most frequent movie duration in the 1990s
# Tìm thời lượng phim thường xuyên nhất trong thập niên 90s
unique, counts = np.unique(durations, return_counts=True)
most_frequent_duration = unique[np.argmax(counts)]
print(f"Thời lượng phim thường xuyên nhất: {most_frequent_duration}")
```

```python
# Get column 'genre' to find count of action-genre
# Lấy và chuyển cột 'genre' về dạng mảng numpy (cho việc tính toán trên numpy đơn giản hơn) để đếm các phim hành động 
genres = netflix90s_df['genre'].to_numpy()
# Set condition variables 
# Đặt các đối số phù hợp điều kiện
is_short = durations < 90
is_action = netflix90s_df['genre'].str.contains('Action', case=False, na=False)
# Count short_action_movies
short_action_movies = np.sum(is_short & is_action)
print(f"Số short action movies released trong thập niên 90s: {short_action_movies}")
```

```python
# Save the results as required
# Trả về kết quả phù hợp kiểu dữ liệu của yêu cầu đề bài
duration = int(most_frequent_duration)
short_movie_count = int(short_action_movies)
```

```python
# Plot the distribution of movie durations in the 1990s
# Đồ thị phân bổ thời lượng phim 90s, sử dụng histogram vói bins = 30 để dễ quan sát 
plt.figure(figsize=(10, 6))
plt.hist(durations, bins=30, edgecolor='k', alpha=0.7)
plt.title('Phân bổ thời lượng phim trong thập niên 90s')
plt.xlabel('Thời lượng phim (phút)')
plt.ylabel('Số lượng phim')
plt.grid(True)
plt.show()
```

**Nhận xét**: Ta có thể thấy thời lượng phim thường xuyên nhất đã được tính ở trên là 94 và ở đồ thị trên cột cao nhất chính là cột 94

```python
# Plot the count of movies by genre in the 1990s
# Thống kê xếp hạng các thể loại phim trong thập niên 90s
genre_counts = netflix90s_df['genre'].value_counts()
plt.figure(figsize=(12, 8))
plt.barh(genre_counts.index, genre_counts.values, color='skyblue')
plt.title('Xếp hạng các thể loại phim trong 90s')
plt.xlabel('Số phim')
plt.ylabel('Thể loại')
plt.grid(True)
plt.show()
```

```python
# Plot the count of short action movies in the 1990s
# Histogram để quan sát sự phân bổ thời lượng phim thuộc thể loại hành động 90s
short_action_movies_df = netflix90s_df[is_short & is_action]
plt.figure(figsize=(8, 5))
plt.hist(short_action_movies_df['duration'].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title('Phân bổ thời lượng phim hành động 90s (Duration < 90 minutes)')
plt.xlabel('Thời lượng phim (phút)')
plt.ylabel('Số lượng phim')
plt.grid(True)
plt.show()
```

**Nhận xét**: tổng số cột trên đồ thị là 7 hoàn toàn tương thích với giá trị đã tính toán ở trên
